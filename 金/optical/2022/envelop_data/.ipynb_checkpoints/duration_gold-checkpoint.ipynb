```python
import numpy as np
import matplotlib.pyplot as plt
import OctopusTools.Basic as cx
```

```python
path = "/data117/chen/jupyter/2nonlinearity/au/envelop_data/"
fs = 14 
dt = 0.08

def env1(x,y, pulse_duration):
    #  exp attenuation
    gamma = 0.005
    envy = max(y)*np.exp(-gamma * (x - pulse_duration) ) * np.heaviside(x - pulse_duration-0.01,1)   \
           +    max(y)* np.heaviside( pulse_duration - x ,1)
    newy = envy  * y / max(envy)
    
    return(newy, envy)

def env2(x,y, pulse_duration):
    #  sin2 attenuation
    tail = 1000    
    envy =   max(y) * (np.heaviside(x - pulse_duration, 1)   * \
             np.square( np.sin( (x - pulse_duration)  *  np.pi / tail /2    + np.pi/2)) + \
             np.heaviside(pulse_duration - x - 0.01, 1) ) *  np.heaviside( pulse_duration + tail - x, 1)
    newy = envy * y / max(envy)
    return(newy, envy)

def plot_env():

    time_ext, A_ext = cx.ReadData(path + "A%s/laser" % 0.1 ,1,2)
    start = 0 
    end = int(5000 / dt)

    time_ext, A_ext = cx.shortten(time_ext, A_ext, start, end)

    newy1, envy1 = env1(time_ext, A_ext, 2000)
    newy2, envy2 = env2(time_ext, A_ext, 2000)
    
    plt.figure(figsize = (16,4))
    
    plt.subplot(1,2,1)
    plt.plot(time_ext , newy1)
    plt.plot(time_ext , envy1, label = "$y = e^{-0.005t}$" )
    plt.legend(fontsize = fs, loc = 1)
    
    plt.subplot(1,2,2)
    plt.plot(time_ext , newy2)
    plt.plot(time_ext , envy2, label = "$sin^2( t )$")
    plt.xlabel("Propagation time [au]", fontsize = fs)
    plt.ylabel("Envlope y ", fontsize = fs)
    
    plt.legend(fontsize = fs, loc = 1)
    plt.xlabel("Propagation time [au]", fontsize = fs)


    plt.show()
    return()


def env(x,y, pulse_duration):
    #  sin2 attenuation
    tail = 1000    
    envy =   max(y) * (np.heaviside(x - pulse_duration, 1)   * \
             np.square( np.sin( (x - pulse_duration)  *  np.pi / tail /2    + np.pi/2)) + \
             np.heaviside(pulse_duration - x - 0.01, 1) ) *  np.heaviside( pulse_duration + tail - x, 1)
    newy = envy * y / max(envy)
    return(newy)
```

```python
plot_env()
```

```python
pulse_duration = 500
head = 1000
tail = 1000

plt.figure(figsize = (14,3)) 

for a in [0.1]:
    time_ext, A_ext = cx.ReadData(path + "A%s/laser" % a ,1,2)
    time_ind, A_ind = cx.ReadData(path + "A%s/gauge_field" % a,1,2)   
    
    start = 0 
    end = len(A_ind)
    
    time_ext, A_ext = cx.shortten(time_ext, A_ext, start, end)
    time_ind, A_ind = cx.shortten(time_ind, A_ind, start, end)    
        
    E_ext = cx.A2E(time_ext, A_ext)   # V/Angstrom
    E_ind = cx.A2E(time_ind, A_ind)
    
    ######## Add the envelope ######################
    A_ext = env(time_ext, A_ext, pulse_duration+head)
    A_ind = env(time_ind, A_ind, pulse_duration+head)
    E_ext = env(time_ext, E_ext, pulse_duration+head)
    E_ind = env(time_ind, E_ind, pulse_duration+head)
    
    
    plt.subplot(1,2,1)
    
    plt.plot(time_ext, A_ext,color = "blue", label = "Ext")
    plt.legend(fontsize = fs, loc = 1)
    plt.xlabel("Propagation time t [$\hbar/Ha$]", fontsize = fs)
    plt.ylabel("Gauge field A [$Ha$]", fontsize = fs)
        
    plt.subplot(1,2,2)
#     plt.plot(time_ext * 0.02419, E_ext, color = "blue")
    plt.plot(time_ind, A_ind, color = "red", label = "Ind")
    plt.legend(fontsize = fs, loc = 1)
#     plt.legend(fontsize = fs, loc = 1)
#     plt.xlabel("Propagation time t [$fs$]", fontsize = fs)
#     plt.ylabel("Electric field E [$V/\mathring{A}$]", fontsize = fs)
    
plt.title("Pulse duration = %s [au]" % pulse_duration, fontsize = fs)       
plt.show()
    



```
